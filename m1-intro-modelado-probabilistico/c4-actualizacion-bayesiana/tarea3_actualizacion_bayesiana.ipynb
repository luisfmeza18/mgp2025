## 1. Actualización Bayesiana discreta

### 1.1.
Tenemos tres tipos de monedas con diferentes probabilidades de caer cara:

- Tipo A: moneda justa, con probabilidad de cara de $0.5$.
- Tipo B: moneda cargada, con probabilidad de cara de $0.6$.
- Tipo C: moneda cargada, con probabilidad de cara de $0.9$.

Tenemos un recipiente con cinco monedas, dos (2) de tipo A, dos (2) de tipo B y una (1) de tipo C. Seleccionamos una moneda del recipiente de forma aleatoria, y sin mirarla, la tiramos y obtenemos **SELLO** ¿Cuál es la probabilidad de que la moneda sea tipo A?, ¿O tipo B?, ¿O tipo C?

#### Planteamiento

Se tienen los siguientes datos del problema:

$$
\text{Moneda}: M = \{\text{Sello}, \text{Cara}\}
\\
\text{Tipo}: T = \{A, B, C\}
$$

$$ p(M=\text{Sello}|T=A) = 0.5, \qquad p(T=A) = 0.4 $$
$$ p(M=\text{Sello}|T=B) = 0.4, \qquad p(T=B) = 0.4 $$ 
$$ p(M=\text{Sello}|T=C) = 0.1, \qquad p(T=C) = 0.2 $$


La probabilidad de que esta moneda sea del tipo $A$, $B$ o $C$ se puede calcular utilizando la **Regla de Bayes**:

$$p(T=A|M=\text{Sello})=\frac{p(M=\text{Sello}|T=A) \cdot p(T=A)}{p(M=\text{Sello})}$$
$$p(T=B|M=\text{Sello})=\frac{p(M=\text{Sello}|T=B) \cdot p(T=A)}{p(M=\text{Sello})}$$
$$p(T=C|M=\text{Sello})=\frac{p(M=\text{Sello}|T=C) \cdot p(T=A)}{p(M=\text{Sello})}$$

Ahora, la probabilidad de obtener un Sello puede calcularse por medio de la **Regla de la Probabilidad Total**:

$$p(M=\text{Sello})=(0.5)(0.4)+(0.4)(0.4)+(0.1)(0.2)=0.38$$

Finalmente, sustituimos los valores en cada ecuación y se llega a las siguientes probabilidades:

$$p(T=A|M=\text{Sello})=\frac{(0.5)(0.4)}{0.38}=\boxed{0.5263}$$
$$p(T=B|M=\text{Sello})=\frac{(0.4)(0.4)}{0.38}=\boxed{0.4210}$$
$$p(T=C|M=\text{Sello})=\frac{(0.1)(0.2)}{0.38}=\boxed{0.0526}$$

### 1.2.

Supongamos que elegimos una moneda como en el ejemplo anterior. La tiramos y obtenemos **SELLO**. Tomamos la misma moneda y la tiramos, y volvemos a obtener **SELLO**. ¿Cuál es la probabilidad de que la moneda sea tipo A?, ¿O tipo B?, ¿O tipo C?

#### Planteamiento

Para resolver esta pregunta se utilizará el concepto de actualización bayesiana y se define lo siguiente:

**Probabilidades previas:**
$$p(T=A) = 0.4, \qquad p(T=B) = 0.4, \qquad p(T=C) = 0.2$$

**Verosimilitudes:**
$$p(M=\text{Sello}|T=A) = 0.5, \qquad p(M=\text{Sello}|T=B) = 0.4, \qquad p(M=\text{Sello}|T=C) = 0.1 $$

**Probabilidades posteriores:**
$$p(T=A|M=\text{Sello}), \qquad p(T=B|M=\text{Sello}), \qquad p(T=C|M=\text{Sello}) $$



Se construye una tabla de actualización bayesiana:

| Hipótesis | Previa | Verosimilitud              | Numerador de Bayes                      | Posterior                 |
| --------- | ------ | ---------------            | --------------------                    | ---------------           |
| $T$       | $p(T)$ | $p(M = \text{Sello} \| T)$ | $p(M = \text{Sello} \| T) \cdot p(T)$   | $p(T \| M = \text{Sello})$|
|           |        |                            |                                         |                           |
| A         | 0.4    | 0.5                        | 0.2                                     | 0.5263                    |
| B         | 0.4    | 0.4                        | 0.16                                    | 0.4210                    |
| C         | 0.2    | 0.1                        | 0.02                                    | 0.0526                    ||
| Total     | 1      | N.A.                       | $p(M = \text{Sello}) = $ 0.38           | 1                         |

En este contexto, esta tabla equivale al caso anterior y se procede a hacer la siguiente iteración pero ahora considerando como previa a la posterior del experimento pasado y se puede calcular la **Probabilidad Predictiva Posterior**:

| Hipótesis | Previa | Verosimilitud              | Numerador de Bayes                      | Posterior                 |
| --------- | ------ | ---------------            | --------------------                    | ---------------           |
| $T$       | $p(T)$ | $p(M = \text{Sello} \| T)$ | $p(M = \text{Sello} \| T) \cdot p(T)$   | $p(T \| M = \text{Sello})$|
|           |        |                            |                                         |                           |
| A         | 0.5263 | 0.5                        | 0.2631                                  | 0.6024                    |
| B         | 0.4210 | 0.4                        | 0.1684                                  | 0.3855                    |
| C         | 0.0526 | 0.1                        | 0.0053                                  | 0.0120                    |
|           |        |                            |                                         |                           |
| Total     | 1      | N.A.                       | $p(M = \text{Sello}) = $ 0.4368         | 1                         |

$$p(T=A|M=\{\text{Sello},\text{Sello}\}) = \boxed{0.6024}$$

## 2. 

Supongamos que tenemos una moneda con probabilidad desconocida $\theta$ de caer en cara. El valor de $\theta$ es aleatorio y podría ser cualquier cosa entre 0 y 1.

Para este ejemplo, supongamos que $\theta$ sigue una pdf previa tipo Beta(3, 1) $p(\theta) = \frac{1}{\Beta(3, 1)} \theta^2$. 

> La expresión constante $\Beta(3, 1)$ se puede expresar en términos de la función $\Gamma$ (generalización del factorial) como

> $$
\Beta(3, 1) = \frac{\Gamma(3) \Gamma(1)}{\Gamma(3 + 1)} = \frac{2! \cdot 0!}{3!} = \frac{1}{3}.
$$

Supongamos que observamos en tres tiros de la moneda, la secuencia cara-cara-sello. Calcule la pdf posterior para $\theta$:

#### Planteamiento

En este caso, se igual forma se puede resolver utilizando una tabla de actualización bayesiana, donde la distribución previa es continua y las verosimilitudes son discretas.

**Probabilidad previa:**
$$p(\theta) = 3\theta^2, \qquad \theta \sim \text{Beta}(3, 1)$$

**Verosimilitud:**
$$\text{Moneda}: \qquad x = \{0, 1\}$$
$$p(x=1 | \theta) = \theta, \qquad p(x=0 | \theta) = 1 - \theta$$

**Numerador de Bayes :**
$$p(x| \theta) \cdot p(\theta) $$

| Hipótesis | Rango          | Previa                                                 | Verosimilitud                  | Numerador de Bayes                                         | Posterior                  |
| --------- | -------------- | ------------------------------------------             | --------------------------     | ---------------------------------------------------------- | -------------------------- |
| $\theta$  | rango $\theta$ | $p(\theta)$                                            | $p(x = \{1, 1, 0\} \| \theta)$ | $p(x = \{1, 1, 0\} \| \theta) \cdot p(\theta)$             | $p(\theta \| x = \{1, 1, 0\})$ |
|           |                |                                                        |                                |                                                            |                            |
| $\theta$  | [0, 1]         | $3\theta^2$                                            | $\theta^2(1-\theta)$           | $3 \theta^4 (1-\theta)$                                    | $30 \theta^4 (1-\theta)$   |
|           |                |                                                        |                                |                                                            |                            |
| Total     | [0, 1]         | $\int_{0}^{1} 3 \theta^2 \text{d}\theta = 1$           | N.A.                           | $\int_{0}^{1} 3 \theta^4 (1-\theta) \text{d}\theta = 1/10$ | $\int_{0}^{1} 30 \theta^4 (1-\theta) \text{d}\theta = 1$

Se pueden graficar la distribución previa y posterior para observar la actualización de las probabilidades

```python
from matplotlib import pyplot as plt
import numpy as np

theta = np.linspace(0, 1, 1001)

prior = 3*theta**2
posterior = 30*theta**4*(1 - theta)

plt.plot(theta, prior, label=r'Previa $p(\theta)$')
plt.plot(theta, posterior, label=r'Posterior $p(\theta|x=\{1,1,0\})$')
plt.legend()
plt.xlabel('$\\theta$')
```

También se puede observar la comparación de la distribución posterior y el numerador de Bayes:

```python
num_bayes = 3*theta**4*(1 - theta)
plt.subplot(2, 1, 1)
plt.plot(theta, posterior, label=r'Posterior $p(\theta|x=\{1,1,0\})$')
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(theta, num_bayes, label=r'Numerador de Bayes $p(x=\{1,1,0\}|\theta)\cdot p(\theta)$')
plt.legend()
```
